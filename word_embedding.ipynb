```python
import gensim.downloader as api
```

```python
# it give the list of embedding models inside gensim
info = api.info()
for model_name, model_data in sorted(info['models'].items()):
  print('%s (%d records): %s' %(model_name, model_data.get('num_records', -1), model_data['description'][:40] + '...'))
```

```python
wv = api.load("word2vec-google-news-300") #word2vec-google-news-300 is a pre-trained word embedding model that represents words as 300-dimensional vectors
```

```python
glove = api.load("glove-twitter-50")
```

```python
fastex = api.load("fasttext-wiki-news-subwords-300")
```

Pretrained Word2Vec fails on unseen words

```python
wv.most_similar("teacher", topn=5)
```

```python
wv.similarity("python", "java")
```

```python
wv.similarity("coffee", "tea")
```

```python
#Word2Vec learns word meanings by predicting words from their local context.
wv.most_similar('time')
```

```python
#GloVe captures semantic relationships using statistics.
glove.most_similar('tea')
```

```python
#FastText represents a word as a bag of character n-grams.
fastex.most_similar('tea')
```

```python
fastex.most_similar("teaching", topn=5)
```

```python
print(wv.distance('tea', 'coffee'))
print(wv.distance('tea', 'pea'))
```

```python
glove.distance('tea', 'pea')
```

```python
print(fastex.distance('tea', 'pea'))
print(fastex.distance('tea', 'coffee'))
```

```python
glove.most_similar(
    positive=["king", "woman"],
    negative=["man"]
)
```

```python
wv.most_similar_cosmul(positive= ['king','women'],negative = ['man'])
```

```python
wv.most_similar_cosmul(positive= ['restaurant','cocktail'], negative = ['dinner'])
```

```python
glove.most_similar(positive= ['restaurant','cocktail'], negative = ['dinner'])
```

✅ GloVe → clean text, large corpus, semantic analysis

✅ FastText → noisy text, Indian languages, social media, misspellings
